````python
# Train Validation Test Split
from sklearn.model_selection import train_test_split
print(merged_df.shape)
train_df,val_df=train_test_split(merged_df.sort_values('Date'),test_size=0.25,shuffle=False)
print(train_df.shape,val_df.shape, merged_test_df.shape)
````

````python
#numeric categorical cols
numerical_cols=['Store','Day','Month','Year']
categorical_cols=['DayOfWeek','StateHoliday','Promo', 'StoreType','StateHoliday', 'Assortment']
# numeric_cols, categorical_cols
````

##Fill NA


````python
from sklearn.impute import SimpleImputer

si=SimpleImputer(strategy='mean')
si.fit(train_input[numerical_cols])
````

````python
train_df[numerical_cols]=si.transform(train_df[numerical_cols])
val_df[numerical_cols]=si.transform(val_df[numerical_cols])
merged_test_df[numerical_cols]=si.transform(merged_test_df[numerical_cols])
````

````python
[train_df.Date.min(), train_df.Date.max()],[val_df.Date.min(), val_df.Date.max()]
````

#Scale (1,0)


````python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms.fit(train_input[numerical_cols])

train_df[numerical_cols]=mms.transform(train_df[numerical_cols])
val_df[numerical_cols]=mms.transform(val_df[numerical_cols])
merged_test_df[numerical_cols]=mms.transform(merged_test_df[numerical_cols])
````

##Encode categorical variables

````python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse=False, handle_unknown='ignore')
ohe.fit(train_df[categorical_cols].astype('str'))
encoded_cols=ohe.get_feature_names_out(categorical_cols)
````

````python
train_df[encoded_cols]=ohe.transform(train_df[categorical_cols].astype('str'))
val_df[encoded_cols]=ohe.transform(val_df[categorical_cols].astype('str'))
merged_test_df[encoded_cols]=ohe.transform(merged_test_df[categorical_cols].astype('str'))
````

````python
X_train=pd.concat([train_df[encoded_cols], train_df[numerical_cols]], axis=1)
X_val=pd.concat([val_df[encoded_cols], val_df[numerical_cols]], axis=1)
X_test=pd.concat([merged_test_df[encoded_cols], merged_test_df[numerical_cols]], axis=1)


y_train=train_df.Sales
y_val=val_df.Sales
````

````python
##mean
from sklearn.metrics import mean_squared_error
def return_mean(inputs):
  return np.full(len(inputs), merged_df.Sales.mean())

train_preds=return_mean(X_train)
#Evaluate
mean_squared_error(y_train,train_preds, squared=False)
````

````python
##random guess
hi, lo=merged_df.Sales.max(),merged_df.Sales.min()
def random_guess(inputs):
  return np.random.random(len(inputs))*(hi-lo)+lo

train_preds=random_guess(X_train)
#Evaluate
mean_squared_error(y_train,train_preds, squared=False)
````

````python
#average same month last year
import datetime


# Extract the current month and year
def smly(inputs):
  previous_year = merged_df['Year'] - 1
  current_month = merged_df.Month

  # Filter the sales data for the previous year's current month
  previous_year_sales = merged_df[(merged_df['Year'] == previous_year) & (merged_df['Month'] == current_month)]

  # Calculate the average sales for the previous year's current month
  average_sales = previous_year_sales['Sales'].mean()

  # Check if average sales are 0
  if average_sales == 0:
      result = 0
  else:
      result = average_sales
  return np.full(len(inputs),result)

smly(X_train)
````

````python
#General Model For Linear Model
model=LinearRegression()
def try_model(model):
  #model fit
  model.fit(X_train,y_train)

  #Generate prediction
  y_pred=model.predict(X_train)
  val_pred=model.predict(X_val)

  # compute RMSE
  train_rmse=mean_squared_error(y_train,y_pred, squared=False)
  val_rmse=mean_squared_error(y_val,val_pred,squared=False)

  return train_rmse,val_rmse
````
